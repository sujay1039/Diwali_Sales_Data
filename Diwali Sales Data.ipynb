GENDER EDA:

```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender', data = df , palette = 'pastel' , hue = 'Gender')
for bars in ax.containers:
     ax.bar_label(bars)
plt.show()
```

```python
sales_gen = df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x = 'Gender', y = 'Amount' , data = sales_gen , hue = 'Gender', palette = 'pastel')
plt.show()
```

From above graphs we can see that most of the buyers are females and the purchasing power of females are greater than male.

AGE EDA:

```python
df.columns
```

```python
ax = sns.countplot(x ='Age Group' , data = df , palette = 'pastel' , hue = 'Gender')
for bars in ax.containers:
     ax.bar_label(bars)
plt.show()
```

```python
#Total Amount V\S Age Group
sales_age = df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x = 'Age Group' , y = 'Amount' , data = sales_age , palette = 'pastel' , hue = 'Age Group')
plt.show()
```

From above graphs we can that most the buyers are of age group between 26-35 yrs female.

STATE EDA:

```python
df.columns
```

```python
#total number of orders from top 10 states 
sales_state = df.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by = 'Orders' , ascending = False).head(10)
sns.barplot(x = 'State' , y = 'Orders' , data = sales_state , palette = 'pastel' , hue = 'State')
sns.set(rc = {'figure.figsize':(20,5)})
plt.show()
```

```python
#total amount/sales from top 10 states
sales_state = df.groupby(['State'],as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = False).head(10)
sns.barplot(x = 'State' , y = 'Amount' , data = sales_state , palette = 'pastel' , hue = 'State')
sns.set(rc = {'figure.figsize':(20,5)})
plt.show()
```

Form the above graphs we can see that unexpectedly most the orders are from Uttar Pradesh, Maharastra and karnataka respectively but total sales/amount is from UP, Karnataka and then Maharastra
